````python
import pandas as pd
import matplotlib.pyplot as plt

# Load each file into a separate DataFrame, ensuring that data is read as numeric
files = ["17058_goodss_H.cat", "17058_goodss_I.cat", "17058_goodss_J.cat", "17058_goodss_V.cat"]
labels = ["H", "I", "J", "V"]
data = {}
for file, label in zip(files, labels):
    df = pd.read_csv(file, sep='\s+', names=["vector_dist_arcsec", "vector_dist_kpc", "sb", "sb_error"])
    # Convert columns to numeric to avoid any type-related issues
    df["vector_dist_kpc"] = pd.to_numeric(df["vector_dist_kpc"], errors='coerce')
    df["sb"] = pd.to_numeric(df["sb"], errors='coerce')
    df["sb_error"] = pd.to_numeric(df["sb_error"], errors='coerce')
    data[label] = df.dropna()  # Drop any rows with non-numeric values

# Create a 2x2 grid for the individual plots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs = axs.flatten()  # Flatten to easily iterate over

# Plot each profile with error bars
for i, (label, df) in enumerate(data.items()):
    axs[i].errorbar(df['vector_dist_kpc'], df['sb'], yerr=df['sb_error'], fmt='o-', label=f"{label}-band")
    axs[i].invert_yaxis()  # Invert y-axis to have smaller numbers at the top
    axs[i].set_xlabel("Distance (kpc)")
    axs[i].set_ylabel("Surface Brightness")
    axs[i].set_title(f"{label}-band")
    axs[i].legend()

# Adjust layout
plt.tight_layout()

# Create the combined plot with all bands
plt.figure(figsize=(8, 6))
for label, df in data.items():
    plt.errorbar(df['vector_dist_kpc'], df['sb'], yerr=df['sb_error'], fmt='o-', label=f"{label}-band")

# Customize the combined plot
plt.gca().invert_yaxis()  # Invert y-axis for combined plot
plt.xlabel("Distance (kpc)")
plt.ylabel("Surface Brightness")
plt.title("Galaxy Surface Brightness Luminosity Profiles")
plt.legend()
plt.show()
````
