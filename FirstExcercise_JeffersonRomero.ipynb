```python
#Función para la primera tarea.
def FIRST_TASK():
    while True:
        try:
            nn=int(input('Escribir el número de elementos que desee ingresar a la matriz: '))
            if nn<=0:
                print("El valor ingresado no es válido")
                continue
            elif math.sqrt(nn)%1!=0.0:
                print('No se puede crear una matriz con este  número de elementos')
                continue
            else:
                print(f'\nsí puede crearse una matriz con {nn} elementos que tendrá {int(math.sqrt(nn))} columnas y {int(math.sqrt(nn))} filas')
                break   
        except ValueError:
            print("Lo ingresado no tiene el formato correcto")
            continue
    matrix=[]
    fila=[]
    ii=1
    jj=1
    while ii<=nn:
        for elemento in range(nn):
            elemento=(int(random.uniform(0,100)))
            ii=ii+1
            fila.append(elemento)
            if jj==int(math.sqrt(nn)):
                matrix.append(fila)
                fila=[]
                jj=1
            else:
                jj=jj+1
                continue
    print(f'\nLa matriz de {nn} elementos creada es: \n{matrix}')
    print('\n',first_task,'\n')
    while True:
        try:
            aa=int(input(f'Seleccione una fila desde 1 hasta {int(math.sqrt(nn))}: '))
            if aa>=1 and aa<=math.sqrt(nn):
                print(f'\nLa fila seleccionada es {aa} cuyos elementos son: \n{matrix[aa-1]}')
                break
            else:
                print('Lo ingresado no es correcto, intente nuevamente')
                continue
        except ValueError:
            continue
    print(f'\nLa suma de los elementos de la fila {aa} es: {(sum(matrix[aa-1]))} ')
FIRST_TASK()
```

```python
#Función para la tercera tarea.
def THIRD_TASK():
    while True:
        try:
            nn=int(input('Escribir el número de elementos que desee ingresar a la matriz: '))
            if nn<25:
                print("El valor ingresado no es válido, debe ser una matriz de al menos 5x5 (25 elementos)")
                continue
            elif math.sqrt(nn)%1!=0.0:
                print('No se puede crear una matriz con este  número de elementos')
                continue
            else:
                print(f'\nsí puede crearse una matriz con {nn} elementos que tendrá {int(math.sqrt(nn))} columnas y {int(math.sqrt(nn))} filas')
                break   
        except ValueError:
            print("Lo ingresado no tiene el formato correcto\n")
            continue
    print('\n',third_task)
    matrix2=[]
    fila=[]
    ii=1
    jj=1
    while ii<=nn:
        for elemento in range(nn):
            elemento=int(random.uniform(0,100))
            ii=ii+1
            fila.append(elemento)
            if jj==int(math.sqrt(nn)):
                matrix2.append(fila)
                fila=[]
                jj=1
            else:
                jj=jj+1
                continue
    print(f'\nLa matrix de {nn} elementos creada es: \n{matrix2}')
    fila5=matrix2[4]
    print(f'\nLos valores de la 5th fila de la matriz son: \n{fila5}')
    columna5=[] 
    ii=1
    jj=0
    for elemento in range(int(math.sqrt(nn))):
        elemento=matrix2[jj][4]
        ii=ii+1
        jj=jj+1
        columna5.append(elemento)
    print(f'\nLos valores de la 5th columna de la matriz son: \n{columna5}')
    suma5=[]
    gg=0
    for gg in range (int(math.sqrt(nn))):
        suma5.append(fila5[gg]+columna5[gg])
        gg=gg+1
    print(f'\nLa suma de cada elemento de la 5th fila más cada elemento de la 5th columna es: \n{suma5}')
THIRD_TASK()
```
