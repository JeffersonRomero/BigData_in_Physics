```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_and_clean_data(filename):
    df = pd.read_csv(filename, sep='\s+')
    df.dropna(inplace=True)
    return df
```

```python
df = load_and_clean_data('archivo_tabulado.cat')
dfH = load_and_clean_data('17058_goodss_H.cat')
dfI = load_and_clean_data('17058_goodss_I.cat')
dfJ = load_and_clean_data('17058_goodss_J.cat')
dfV = load_and_clean_data('17058_goodss_V.cat')
```

```python
def plot_histogram(data, title, xlabel, ylabel, xlim=(0.1, 3), color='gray', edgecolor='black'):
    plt.hist(data, color=color, edgecolor=edgecolor)
    plt.xlim(left=xlim[0], right=xlim[1])
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
plot_histogram(df['z'], 'Redshift Distribution', r'Redshift $(z)$', 'Frequency')
```

```python
def plot_mass_vs_radius(df, mass_column='mass', radius_column='re', n_column='n'):
    filtered_data = df[(df[mass_column] > 3e9) & (df[radius_column] > 0)]
    plt.scatter(filtered_data[mass_column][filtered_data[n_column] <= 2.5],
                filtered_data[radius_column][filtered_data[n_column] <= 2.5],
                color='purple', label='n <= 2.5', marker='s')
    plt.scatter(filtered_data[mass_column][filtered_data[n_column] > 2.5],
                filtered_data[radius_column][filtered_data[n_column] > 2.5],
                color='green', label='n > 2.5', alpha=0.5)
    plt.axvline(5e10, color='red', linestyle='dashed', label='$5 \\times 10^{10} M_{\\odot}$')
    plt.axvline(1e11, color='blue', linestyle='dashed', label='$10^{11} M_{\\odot}$')
    plt.xscale('log')
    plt.xlabel('Mass ($M_{\\odot}$)')
    plt.ylabel(r'Effective Radius $(re)$')
    plt.title('Mass vs Effective Radius')
    plt.legend()
    plt.grid(False)
    plt.show()
```

```python
plot_mass_vs_radius(df)
```
